```python
#lets import the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
kidnap = pd.read_csv('39_Specific_purpose_of_kidnapping_and_abduction.csv')
kidnap.head()
```

```python
kidnap.info()
```

```python
#lets plot the reasons for kidnapping
fig, ax = plt.subplots(figsize = (10,10))
sns.countplot(kidnap['Sub_Group_Name'])
plt.title('Reasons for kidnapping', fontsize = 16)
plt.xlabel('Reasons', fontsize  = 14)
plt.ylabel('Count', fontsize =  14)
plt.xticks(rotation = 90);
```
